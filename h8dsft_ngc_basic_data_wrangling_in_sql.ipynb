```python
#code answers


# in austin_bikeshare dataset, there are 2 tables, bikeshare_stations and bikeshare_trips
df = client.query('''
SELECT COUNT(*)
FROM `bigquery-public-data.austin_bikeshare.__TABLES__`
''').to_dataframe()
df
```

```python
# in bikeshare_stations table, there are 102 entries, with 14 columns and within them there are 7 object/strings data types
df = client.query('''
SELECT *
FROM `bigquery-public-data.austin_bikeshare.bikeshare_stations`
''').to_dataframe ()
df.info()
```

```python
#Retrieve all trips where the bike_id is not null and the duration is greater than 1000 minutes.

df = client.query('''
SELECT *
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
WHERE bike_id IS NOT NULL
  AND duration_minutes > 1000;
''').to_dataframe()
df
```

```python
#Filter trips where the subscriber_type is "Walk Up" and the end_station_name  is "Stolen"

df = client.query('''
SELECT *
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
WHERE subscriber_type = "Walk Up"
  AND end_station_name = "Stolen";
''').to_dataframe()
df
```
